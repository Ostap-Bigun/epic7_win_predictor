```python
class Battle(BaseModel):
    battleCompletedate: datetime
    teamBettleInfoenemy: str
    prebanListEnemy: str
    teamBettleInfo: str
    prebanList: str
    iswin: int

class ResultBody(BaseModel):
    nick_no: int
    battle_list: List[Battle]

class Response(BaseModel):
    result_body: ResultBody
```

```python
class MyTeam(BaseModel):
    pick_order: int
    hero_code: str
    attack_damage: int
    receive_damage: float
    recovery: int
    mvp_point: int
    artifact: str
    equip: list
    respawn: int
    mvp: int
    kill_count: int 
    attribute_cd: str
    job_cd: str

class BattleInfo(BaseModel):
    my_team: List[MyTeam]
```

```python
# Extract hero parameters from the string and add them to the corresponding lists - Right Side
enemy_team = []
for i in range(len(my_cool_response.result_body.battle_list)):
    team_enemy_json = my_cool_response.result_body.battle_list[i].teamBettleInfoenemy
    team_enemy_data = BattleInfo.model_validate_json('{' + team_enemy_json + '}')
    my_team_enemy = team_enemy_data.my_team
    enemy_team.append(my_team_enemy)

pick_order = []
hero_code = []
attack_damage = []
receive_damage = []
recovery = []
mvp_point = []
artifact = []
equip = []
respawn= []
mvp= []
kill_count= []
attribute_cd= []
job_cd= []
    
for i in range(len(enemy_team)):
    for j in range(len(enemy_team[i])):
        pick_order.append(enemy_team[i][j].pick_order)
        hero_code.append(enemy_team[i][j].hero_code)
        attack_damage.append(enemy_team[i][j].attack_damage)
        receive_damage.append(enemy_team[i][j].receive_damage)
        recovery.append(enemy_team[i][j].recovery)
        mvp_point.append(enemy_team[i][j].mvp_point)
        artifact.append(enemy_team[i][j].artifact)
        equip.append(enemy_team[i][j].equip)
        respawn.append(enemy_team[i][j].respawn)
        mvp.append(enemy_team[i][j].mvp)
        kill_count.append(enemy_team[i][j].kill_count)
        attribute_cd.append(enemy_team[i][j].attribute_cd)
        job_cd.append(enemy_team[i][j].job_cd)
    
# Create a df
right_side_df = pd.DataFrame({
    'pick_order': pick_order, 'hero_code': hero_code, 'attack_damage': attack_damage, 
    'receive_damage': receive_damage, 'recovery': recovery, 'mvp_point': mvp_point,
    'artifact': artifact, 'equip': equip, 'respawn': respawn, 'mvp': mvp, 
    'kill_count': kill_count, 'attribute_cd': attribute_cd, 'job_cd': job_cd,
    'preban_right': np.nan, 'is_win': np.nan})

#Extract info on prebans and wins
preban_right = []
is_win = []
for i in range(len(my_cool_response.result_body.battle_list)):
    preban_right.append(my_cool_response.result_body.battle_list[i].prebanListEnemy)
    is_win.append(my_cool_response.result_body.battle_list[i].iswin)
        

```
