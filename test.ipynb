```python
import requests
import pandas as pd
import numpy as np
from pydantic import BaseModel, ValidationError
from typing import List
from datetime import datetime

import json
```

```python
class Player(BaseModel):
    nick_no: int
    world_code: str

class MyTeam(BaseModel):
    pick_order: int
    hero_code: str
    attack_damage: int
    receive_damage: float
    recovery: int
    mvp_point: int
    artifact: str
    equip: list
    respawn: int
    mvp: int
    kill_count: int
    attribute_cd: str
    job_cd: str


class TeamBettleInfoenemy(BaseModel):
    my_team: List[MyTeam]


class Battle(BaseModel):
    battleCompletedate: datetime
    teamBettleInfoenemy: str
    prebanListEnemy: str
    teamBettleInfo: str
    prebanList: str
    iswin: int


class ResultBody(BaseModel):
    nick_no: int
    battle_list: List[Battle]


class Response(BaseModel):
    result_body: ResultBody
```

```python
def get_players():
    players = [
        Player(nick_no=56302454, world_code="world_kor"),
        Player(nick_no=71212252, world_code="world_asia"),
    ]
    return players

def get_player_battles(player: Player) -> Response:
    try:
        resp = requests.post(
            url=f"https://epic7.gg.onstove.com/gameApi/getBattleList?nick_no={player.nick_no}&world_code={player.world_code}&lang=en&season_code=",
            headers={
                "Accept-Language": "en-US",
                "Accept": "application/json, text/plain, */*",
            },
        )

        if resp.ok:
            response = Response.model_validate(resp.json())
            for i in range(len(response.result_body.battle_list)):
                response.result_body.battle_list[i].teamBettleInfoenemy = (
                    TeamBettleInfoenemy.model_validate_json("{" + response.result_body.battle_list[i].teamBettleInfoenemy+ "}")
                )
            return response
        else:
            resp.raise_for_status()
    except Exception as e:
        print(e)

def make_df(pbis: List[Response]) -> pd.DataFrame:
    # enemy_team = []
    # pick_order = []
    # hero_code = []
    # attack_damage = []
    # receive_damage = []
    # recovery = []
    # mvp_point = []
    # artifact = []
    # equip = []
    # respawn= []
    # mvp= []
    # kill_count= []
    # attribute_cd= []
    # job_cd= []
    # for pbi in pbis:
    #     for i in range(len(pbi.result_body.battle_list)):
    #         enemy_team.append(pbi.result_body.battle_list[i].teamBettleInfoenemy)


    #     for i in range(len(enemy_team)):
    #         for j in range(len(enemy_team[i])):
    #             pick_order.append(enemy_team[i][j].pick_order)
    #             hero_code.append(enemy_team[i][j].hero_code)
    #             attack_damage.append(enemy_team[i][j].attack_damage)
    #             receive_damage.append(enemy_team[i][j].receive_damage)
    #             recovery.append(enemy_team[i][j].recovery)
    #             mvp_point.append(enemy_team[i][j].mvp_point)
    #             artifact.append(enemy_team[i][j].artifact)
    #             equip.append(enemy_team[i][j].equip)
    #             respawn.append(enemy_team[i][j].respawn)
    #             mvp.append(enemy_team[i][j].mvp)
    #             kill_count.append(enemy_team[i][j].kill_count)
    #             attribute_cd.append(enemy_team[i][j].attribute_cd)
    #             job_cd.append(enemy_team[i][j].job_cd)

    # # Create a df
    # right_side_df = pd.DataFrame({
    #     'pick_order': pick_order, 'hero_code': hero_code, 'attack_damage': attack_damage, 
    #     'receive_damage': receive_damage, 'recovery': recovery, 'mvp_point': mvp_point,
    #     'artifact': artifact, 'equip': equip, 'respawn': respawn, 'mvp': mvp, 
    #     'kill_count': kill_count, 'attribute_cd': attribute_cd, 'job_cd': job_cd,
    #     'preban_right': np.nan, 'is_win': np.nan})

    # # Extract info on prebans and wins
    # preban_right = []
    # is_win = []
    # for i in range(len(my_cool_response.result_body.battle_list)):
    #     preban_right.append(my_cool_response.result_body.battle_list[i].prebanListEnemy)
    #     is_win.append(my_cool_response.result_body.battle_list[i].iswin)
    return pd.DataFrame()

def get_data() -> pd.DataFrame:
    pbis = []
    players = get_players()
    for player in players:
        player_battle_info = get_player_battles(player)
        pbis.append(player_battle_info)
        
    print(pbis)
        
    return make_df(pbis)
```
